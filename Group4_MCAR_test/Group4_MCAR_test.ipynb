# **Little MCAR test : Method 1**

Dropping those columns which have missing value more than or equal to 50%

```python
# Drop columns with more than 50% missing values
missing_rate = df_numeric.isnull().mean()
print(missing_rate)
df_filtered = df_numeric.loc[:, missing_rate < 0.5]
df_filtered
```

```python
# Check if there's any numeric data to process
if df_filtered.empty:
    raise ValueError("No numeric columns with less than 50% missing data found in the dataset. Little's MCAR test cannot be performed.")
```

```python
from scipy import stats
```

```python
# Function to perform a simplified MCAR test
def mc_car_test(df):
    # Convert dataframe to numpy array
    data = df.to_numpy()

    # Count missing values
    missing_values = np.isnan(data)

    # If no missing values, return early
    if not np.any(missing_values):
        print("No missing values found. Data is MCAR by definition.")
        return

    # Calculate missing data pattern
    missing_patterns = np.sum(missing_values, axis=1)

    # Calculate statistics for missing data pattern
    unique_patterns, counts = np.unique(missing_patterns, return_counts=True)
    pattern_probabilities = counts / len(missing_patterns)

    # Perform chi-square test to check the pattern distribution
    chi2_stat, p_value = stats.chisquare(pattern_probabilities)

    print("MCAR Test Results:")
    print(f"Chi-square statistic: {chi2_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value > 0.05:
        print("The null hypothesis cannot be rejected. The data is MCAR.")
    else:
        print("The null hypothesis is rejected. The data is not MCAR.")
```

```python
# Run the test on the filtered dataset
mc_car_test(df_filtered)
```

# **Little MCAR test : Method 2**

```python
from scipy.stats import chi2
```

```python
def little_mcar_test(data):
    n = len(data)  # Number of observations
    groups = []

    # Identify columns with missing values and create a mask for them
    for col in data.columns:
        mask = data[col].isnull()
        if mask.any():
            groups.append(mask.astype(int).values.reshape(-1, 1))

    # Raise an error if no missing data is found
    if len(groups) == 0:
        raise ValueError("No missing data found.")

    # Concatenate the group masks into a single matrix
    r = np.concatenate(groups, axis=1)

    # Compute the group statistics
    group_stats = r.T @ r
    m = len(groups)  # Number of groups (columns with missing data)

    # Calculate degrees of freedom
    df = (n - 1) * m

    # Calculate the chi-squared statistic
    chi2_stat = group_stats.trace()

    # Calculate the p-value
    p_value = chi2.sf(chi2_stat, df)

    # Return the results
    return {
        "chi2_stat": chi2_stat,
        "degrees_of_freedom": df,
        "p_value": p_value
    }
```

```python
# Run the MCAR test
result = little_mcar_test(df)

# Print the results
print(f"Chi-square statistic: {result['chi2_stat']}")
print(f"Degrees of freedom: {result['degrees_of_freedom']}")
print(f"P-value: {result['p_value']}")

# Hypothesis test result
if result['p_value'] > 0.05:
  print("The null hypothesis cannot be rejected. The data is MCAR.")
else:
  print("The null hypothesis is rejected. The data is not MCAR.")
```

# **Little MCAR test : Method 3**

Using inbuilt functions

```python
# Create missingness indicators
missingness_indicators = df.isnull().astype(int)
missingness_indicators
```

```python
# Add columns for missingness indicators
df['median_salary_missing'] = missingness_indicators['median_salary']
df['life_satisfaction_missing'] = missingness_indicators['life_satisfaction']
df['recycling_pct_missing'] = missingness_indicators['recycling_pct']
df['population_size_missing'] = missingness_indicators['population_size']
df['number_of_jobs_missing'] = missingness_indicators['number_of_jobs']
df['area_size_missing'] = missingness_indicators['area_size']
df['no_of_houses_missing'] = missingness_indicators['no_of_houses']

# Create a complete-case dataset (remove rows with any missing values)
complete_data = df.dropna()
complete_data
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import logit
```

```python
# Fit logistic regression for the missingness of 'median_salary'
formula = 'median_salary_missing ~ area_size + population_size + number_of_jobs + life_satisfaction + recycling_pct'
model = logit(formula, data=complete_data).fit()
```

```python
# Check if any p-values are significant
if (model.pvalues < 0.05).any():
    print("Reject the null hypothesis: Data may not be MCAR.")
else:
    print("Fail to reject the null hypothesis: Data may be MCAR.")
```

```python
from fancyimpute import KNN, IterativeImputer
```

```python
# Select only numeric columns for imputation
numeric_df = df.select_dtypes(include=[np.number])
numeric_df
```

```python
# Check for missing values
missing_values = numeric_df.isnull().sum()

# Display missing values count for each column
print(missing_values)
```

```python
# Mean Imputation
mean_filled = numeric_df.fillna(numeric_df.mean())
mean_filled
```

```python
# Check for missing values
missing_values = mean_filled.isnull().sum()

# Display missing values count for each column
print(missing_values)
```

```python
# Median Imputation
median_filled = numeric_df.fillna(numeric_df.median())
median_filled
```

```python
# Check for missing values
missing_values = median_filled.isnull().sum()

# Display missing values count for each column
print(missing_values)
```

```python
# Mode Imputation
mode_filled = numeric_df.fillna(numeric_df.mode().iloc[0])
mode_filled
```

```python
# Check for missing values
missing_values = mode_filled.isnull().sum()

# Display missing values count for each column
print(missing_values)
```

```python
# KNN Imputation
# Select numeric columns for imputation
numeric_columns = numeric_df.select_dtypes(include=['number']).columns

# Perform KNN imputation on the selected numeric columns
knn_imputer = KNN(k=5)  # Set the number of neighbors as needed
df[numeric_columns] = knn_imputer.fit_transform(df[numeric_columns])

# Display the imputed DataFrame
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display missing values count for each column
print(missing_values)
```

KNN imputer does not work for categorical data thus we need to encode it first in order to use KNN imputation and then apply it.

```python
# Encode categorical columns
df_encoded = pd.get_dummies(df, drop_first=True)

# Perform imputation on the encoded DataFrame
df_knn_imputed = pd.DataFrame(KNN(k=5).fit_transform(df_encoded), columns=df_encoded.columns)
```

```python
# Check for missing values
missing_values = df_knn_imputed.isnull().sum()

# Display missing values count for each column
print(missing_values)
```

```python
# Perform Missing Forest imputation (Iterative Imputer)
# Select numeric columns for imputation
numeric_columns = df.select_dtypes(include=['number']).columns

# Perform Missing Forest imputation (Iterative Imputer) on numeric columns
missing_forest_imputer = IterativeImputer()
df[numeric_columns] = missing_forest_imputer.fit_transform(df[numeric_columns])

# Display the imputed DataFrame
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Display missing values count for each column
print(missing_values)
```
