*   We first import the necessary libraries: matplotlib.pyplot, seaborn, sklearn.datasets, and pandas.

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

from IPython.display import Image
```



*  We then load the Iris dataset using load_iris() from sklearn.datasets and convert it into a pandas DataFrame for easier manipulation.




```python
# Loading the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```



*   The code then proceeds to create the specified visualizations using the Iris dataset:





1.  Histograms:
* histogram of sepal length, sepal width, petal length, and petal width



```python

# Histograms
plt.figure(figsize=(12, 6))

sns.histplot(iris_df['sepal length (cm)'], kde=True)
```



*   Subplots allow you to compare multiple datasets or variables side by side. This can help in identifying trends, outliers, and patterns that may not be apparent in a single plot.



```python


plt.figure(figsize=(12, 6))

# Plot for sepal length
plt.subplot(2, 2, 1)
sns.histplot(iris_df['sepal length (cm)'], kde=True)
plt.title('Sepal Length')

# Plot for sepal width
plt.subplot(2, 2, 2)
sns.histplot(iris_df['sepal width (cm)'], kde=True)
plt.title('Sepal Width')
```

2. Box plots:
*   Box plots, also known as box-and-whisker plots, are a type of data visualization that provides a graphical summary of the distribution of a dataset.
*  They are particularly useful for comparing the distribution of multiple
datasets or for visualizing the spread and central tendency of a single dataset.



```python

# Box plots
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris_df, orient='h')
plt.title('Box Plots')
plt.show()
```

```python
plt.figure(figsize=(6, 4))  # Adjust the figure size as needed
sns.boxplot(x=iris_df['sepal length (cm)'])
plt.title('Box Plot of Sepal Length')
plt.show()
```

3. Scatter plot:
* Scatter plot of sepal length vs. sepal width.


* Why we use scatter plot:
 Scatter plots help in identifying any patterns or trends in the relationship between the two variables. For example, you can see if there is a linear, quadratic, or no relationship between the variables.

```python
# Scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df)
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.show()
```



4. Pair Plot:
Pair plot showing pairwise relationships between the features.



```python
# Pair plots
plt.figure(figsize=(12, 8))
sns.pairplot(iris_df)
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

5. Heatmap:
* Heatmap showing the correlation matrix of the features.

```python
# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```


6. Violin plot:
* A violin plot is a data visualization technique used to visualize the distribution of a continuous variable across different categories or groups. It is similar to a box plot but provides more information about the distribution of the data

```python
Image(filename='/content/drive/MyDrive/EDA/Lab 7/r3.png')
```

```python
# Violin plots
plt.figure(figsize=(10, 6))
sns.violinplot(data=iris_df, orient='h')
plt.title('Violin Plots')
plt.show()
```

7. Joint plot:
* jointplot in Seaborn is a versatile function that combines several types of plots to visualize the relationship between two variables. It provides a quick way to understand the joint distribution of variables, along with the individual distributions of each variable.

```python
# Joint plots
sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=iris_df, kind='scatter')
plt.suptitle('Joint Plot of Petal Length vs. Petal Width', y=1.02)
plt.show()
```
